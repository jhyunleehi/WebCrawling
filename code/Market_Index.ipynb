```python
import requests
```

```python
url = 'https://finance.naver.com/marketindex/'
```

```python
res = requests.get(url)
```

```python
res.headers['content-type']
```

```python
from bs4 import BeautifulSoup as bts
```

```python
soup = bts(markup = res.text, features = 'html.parser')
```

```python
dir(soup)
```

```python
soup
```

```python
items = soup.select('#exchangeList > li > a > div > span.value')
len(items)
```

```python
items = soup.select('#exchangeList span.value')
len(items)
```

```python
items
```

```python
result = []
for item in items:
    result.append(item.text)
result
```

```python
result = [item.text for item in items]
result
```

```python
class my_bts(bts):
    def htmlText(self, css):
        items = self.select(css)
        return [item.text for item in items]
```

```python
soup = my_bts(markup = res.text, features = 'html.parser')
```

```python
soup.htmlText(css = '#exchangeList span.value')
```
