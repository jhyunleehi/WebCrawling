```python
corpus = [' '.join(i) for i in corpus]
corpus
```

```python
for change in changes:
    removed = change.replace(' ', '')
    corpus = [i.replace(change, removed) for i in corpus]
```

```python
corpus[0:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
dtm = cv.fit_transform(corpus).toarray()
dtm
```

```python
dtm = pd.DataFrame(data = dtm, columns = cv.get_feature_names_out())
dtm
```
